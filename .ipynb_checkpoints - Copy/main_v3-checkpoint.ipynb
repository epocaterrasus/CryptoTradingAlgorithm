```python
# # import pandas as pd
# # from sklearn.preprocessing import StandardScaler
#define columns to use for PCA
# X = df.drop(columns=["pct_change"])
close = df['close']
df = df.drop(columns=['close'])
print(close.shape)
print(df.shape)

#define scaler
scaler = StandardScaler()

#create copy of DataFrame
scaled_df=df.copy()

#created scaled version of DataFrame
scaled_df=pd.DataFrame(scaler.fit_transform(scaled_df), columns=scaled_df.columns)
```

```python
# Run PCA first time To Determine the Component
# from sklearn.decomposition import PCA
# pca.fit_transform(df)
#define PCA model to use
n_components = 10
pca = PCA(n_components=n_components)
print(pca)

#fit PCA model to data
pcafit = pca.fit_transform(scaled_df) # choose 10 as component number and found 3 component is sufficient for the scree plot.
print(pcafit)
```

```python
# import matplotlib.pyplot as plt
# import numpy as np

PC_values = np.arange(pca.n_components_) + 1

fig, (ax1, ax2) = plt.subplots(1,2,figsize = (16,6))
fig.suptitle('Explained Variance',fontsize=18)
ax1.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
ax2.plot(PC_values, np.cumsum(pca.explained_variance_ratio_), 'o-', linewidth=2,color='red')
ax1.set_xlabel('Principal Component')
ax2.set_xlabel('Number of Principal Components')
ax1.set_ylabel('Variance Explained')
ax2.set_ylabel('Variance Explained')
ax1.xaxis.set_ticks(PC_values)
ax2.xaxis.set_ticks(PC_values)
ax2.yaxis.set_ticks(np.cumsum(pca.explained_variance_ratio_))
ax1.axhline(y=0.08,xmin=0,xmax=3,c="red",linewidth=0.5,zorder=0)
ax2.axhline(y=0.66,xmin=0,xmax=3,c="blue",linewidth=0.5,zorder=0)
plt.show()
```

```python
pca_df = pca_df.to_numpy()
pca_df.shape
```

## Train and test split

```python
train_ratio = 0.8
test_ratio  = 0.2
pca_df_train = pca_df[:int(pca_df.shape[0]*train_ratio),:]
pca_df_test  = pca_df[int(pca_df.shape[0]*train_ratio):,:]
display(pca_df_train.shape)
display(pca_df_test.shape)
```

```python
n_steps = 7
X_train,y_train = split_sequences(pca_df_train,n_steps)
X_test,y_test = split_sequences(pca_df_test,n_steps)
display(X_train.shape)
display(X_test.shape)
```

```python
for i in range(len(X_train)):
    print(f'Timestep {i+1}\n')
    print('Rolling Features:',X_train[i],'\nTarget:',y_train[i])
    print('\n\n')
    if i == 4:
        break
for i in range(len(X_test)):
    print(f'Timestep {i+1}\n')
    print('Rolling Features:',X_test[i],'\nTarget:',y_test[i])
    print('\n\n')
    if i == 4:
        break
```

```python
#transform the tridimensional input tensor in a 2D one by aligning in one row data relevan to the previous steps
n_input = X_train.shape[1] * X_test.shape[2]
n_input
```

```python
X_train = X_train.reshape((X_train.shape[0], n_input))
X_test = X_test.reshape((X_test.shape[0], n_input))
display(X_train.shape)
display(X_test.shape)
```

## Model creation

```python
nn = Sequential([
    Dense(units=2*n_input, input_shape=(n_input,), activation='relu'),
    Dense(units=int(1.5*n_input) ,activation='relu'),
    Dense(units=n_input ,activation='relu'),
    Dense(units=1)
])
```

```python
nn.compile(loss="mse", optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3,clipvalue = 1.0))
epochs = 1000

history = nn.fit(X_train,y_train,epochs=epochs,verbose=1,validation_data=(X_test,y_test))
```

```python
y_train_pred = nn.predict(X_train)
y_pred = nn.predict(X_test)
y_pred
```

```python
history.history.keys()
```

```python
fig,ax = plt.subplots(figsize=(16,5))
fig.suptitle("Model's training and testing loss",fontsize=16)
fig.legend(loc='upper center')
ax2 = ax.twinx()
ax.plot(np.array(range(epochs)),history.history['loss'],label='Training loss')
ax2.plot(np.array(range(epochs)),history.history['val_loss'],color='orange',label='Validation loss')
ax.set_xlim(left=0,right=epochs)
ax.set_ylim(bottom=0,top=1e7)
ax.tick_params(axis='y',labelcolor='blue')
ax.spines['left'].set_color('blue')
ax2.tick_params(axis='y',labelcolor='orange')
ax2.spines['right'].set_color('orange')
```
